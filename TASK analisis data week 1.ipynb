# Proses Data Cleaning

```python
# load new data

df = pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
# Melihat info dataset mengenai missing values

df.info()
```

```python
# Melihat jumlah non null dari data

df.notnull().sum()
```

**Cleaning column Rating**

```python
#menampilkan visualisasi columnnya

df.Rating.plot(kind='hist');
```

```python
#cleaning

val = df.Rating.median()
df['Rating'] = df.Rating.fillna(val)
```

```python
#menampilkan info dataset untuk melihat apakah column Rating sudah ter imputation

df.info()
```

**Cleaning column Type**

```python
df.shape[0]
```

```python
# #menghitung persentase

df['Type'].isna().sum()/df.shape[0]
```

```python
df.Type.value_counts()
```

Note : Column Type memiliki data yang jelas (berisikan tipe aplikasi berbayar dan gratis serta hanya ada 1 missing value, jai dapat dipertahankan

```python
#jumlah data missing

df.Type.isna().sum()
```

```python
#cleaning

val = df.Type.mode().values[0]
df['Type'] = df.Type.fillna(val)
```

```python
#tampilkan proporsi data lagi

df.Type.value_counts()
```

```python
#cek apakah column Type sudah berubah

df.info()
```

Notes: column type dipertahankan karena missing value hanya 1 dan bukan data unique (isi data jelas)

**Cleaning column Content Rating**

Column Contet Rating

```python
df['Content Rating'].isna().sum()/df.shape[0]
```

```python
df['Content Rating'].value_counts()
```

```python
# jumlah data missing

df['Content Rating'].isna().sum()
```

```python
val = df['Content Rating'].mode().values[0]
df['Content Rating'] = df['Content Rating'].fillna(val)
```

```python
# tampilkan proporsi data lagi

df['Content Rating'].value_counts()
```

```python
#cek apkah kolom Content Rating sudah berubah

df.info()
```

Notes : Column Content Rating juga diprtahankan karena missing value hanya 1 dan bukan data unique (isi data jelas)
