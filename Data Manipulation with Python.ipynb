# Pivot Table

Pivot ini merupakan data agar lebih mudah di pahami

```python
df_slice = df[['Survived', 'Pclass', 'PassengerId']]
slice_pivot = df_slice.pivot_table(index=['Survived'], columns=['Pclass'], \
                                    aggfunc=np.size,fill_value=0)
```

```python
slice_pivot
```

```python
df_slice = df[['Survived', 'Sex', 'Pclass', 'PassengerId']]
slice_pivot = df_slice.pivot_table(index=['Survived', 'Sex'], columns=['Pclass'], \
                                    aggfunc=np.size,fill_value=0)
```

```python
slice_pivot
```

# Merge, Join, Concatenate, and Compare

```python
df1 = df[['PassengerId', 'Name', 'Sex']][0:5]
```

```python
df2 = df[['PassengerId', 'Name', 'Sex']][5:10]
```

```python
df_concat = pd.concat([df1, df2])

df_concat
```

```python
df4 = df[['PassengerId', 'Name', 'Sex']][0:5]
```

```python
df5 = df[['PassengerId', 'Age', 'Pclass']][0:5]
```
